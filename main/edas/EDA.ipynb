```python
df['heart_attack_risk'].value_counts()
```

```python
def grafico_de_barras(variavel, data=df, size=(4,4)):    
    plt.figure(figsize=(size))
    sns.set_theme(style='darkgrid', palette= 'deep')
    sns.barplot(data = data[variavel].value_counts().reset_index(),
                x = variavel, 
                y = 'count',
                hue = variavel)
    plt.title(f'contagem: {variavel}')
    plt.ylabel('Quantidade')
    plt.xlabel(variavel)
    plt.show()

```

```python
grafico_de_barras('heart_attack_risk')
```

- Maioria de individuos da amostra NÂO ESTÃO em risco aumentado para ataque cardíaco

#### FATORES INDIVIDUAIS: IDADE, RENDA, PAÍS/CONTINENTE/HEMISFÉRIO

```python
def plotar_distribuição(variavel, data = df, size = (8, 6), bins = 'auto'):
    plt.figure(figsize=(size))
    sns.set_theme(style='darkgrid', palette= 'deep')
    sns.histplot(x = data[variavel], bins = bins)
    plt.title(f'Distribuição das frequências: {variavel}')
    plt.ylabel('Frequência')
    plt.xlabel(variavel)
    plt.show()

def plotar_boxplot_comparativo(variavel, 
                               x = df['heart_attack_risk'], 
                               data = df, 
                               size = (4, 4),
                               rot = 0):
    plt.figure(figsize=(size))
    sns.set_theme(style='darkgrid', palette= 'deep')
    sns.boxplot(data = data,
                y = variavel,
                x = x,
                hue = x)
    plt.title(f'Gráfico comparativo')
    plt.ylabel(variavel)
    plt.xlabel('Risco de Ataque Cardíaco')
    plt.xticks(rotation = (rot))
    plt.legend()
    plt.show()

def comparar_medias(variavel, data=df, size=(4,4)):    
    plt.figure(figsize=(size))
    sns.set_theme(style='darkgrid', palette= 'deep')
    sns.barplot(data = data[variavel].value_counts().reset_index(),
                y = variavel, 
                x = data['heart_attack_risk'],
                hue = data['heart_attack_risk'])
    plt.title(f'contagem: {variavel}')
    plt.ylabel('Quantidade')
    plt.xlabel(variavel)
    plt.show()
```

```python
def barras_groupby(variavel,
                   y = 'count',
                   hue = 'heart_attack_risk',
    size = (6,6)):
        plt.figure(figsize=(size))
        sns.set_theme(style='darkgrid', palette= 'deep')
        sns.barplot(data = df.groupby(variavel)['heart_attack_risk'].value_counts().reset_index(),
                    x = variavel,
                    y = y,
                    hue = hue)
        plt.title(f'Individuos com e sem {variavel} com risco aumentado de ataque cardíaco')
        plt.ylabel('Quantidade')
        plt.xlabel(variavel)
        plt.show()
```

```python
df.groupby('diabetes')['heart_attack_risk'].value_counts()
```

```python
barras_groupby('diabetes')
```

```python
df.groupby('obesity')['heart_attack_risk'].value_counts() 
```

```python
barras_groupby('obesity')
```

```python
df.groupby('smoking')['heart_attack_risk'].value_counts() 
```

```python
barras_groupby('smoking')
```
